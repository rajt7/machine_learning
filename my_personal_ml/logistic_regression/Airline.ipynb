### Handling Missing values

```python
(df.isnull().sum() * 100 / len(df))
```

As 0.30 percent of data is missing in the column 'Arrival Delay in minutes'. We can replace those values.<br>
Creating a strategy to handle missing value<br> 
The duration of delay in minutes in arrival could be approximately guessed by the approximate delay in departure.

```python
df[df['Arrival Delay in Minutes'].isna()]
```

```python
df.loc[df['Arrival Delay in Minutes'].isna(), 'Arrival Delay in Minutes'] = df['Departure Delay in Minutes']
df['Arrival Delay in Minutes'].isnull().sum()
```

### Feature Encoding

Encoding the Customer Type, Type of Travel and Class column in dataset.<br>
Customer Type:- Loyal Customer, disloyal customer<br>
Type of travel:- Business Travel, Personal Travel<br>
Class:- Business, Eco, Eco Plus

```python
# Features
X = df.drop('satisfaction', axis=1)

# Target variable
y = df['satisfaction']
```

```python
X = pd.get_dummies(data=X, drop_first=True)
X
```

```python
# Encoding the target variable
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
le.classes_
```

```python
le.inverse_transform([0])
```

Encoding is like this<br>
1 :- Satified<br>
0 :- dissatisfied

### Splitting the dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_train['Food and drink'].unique()
```

```python
X_train.iloc[:, :-7]
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train.iloc[:, :-7] = sc.fit_transform(X_train.iloc[:, :-7])
X_train = X_train.to_numpy()
```

```python
X_test.iloc[:, :-7] = sc.transform(X_test.iloc[:, :-7])
X_test = X_test.to_numpy()
X_test
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42, max_iter=400, solver='lbfgs')
lr.fit(X_train, y_train)
```

```python
y_pred_prob = lr.predict_proba(X_test)
y_pred_prob
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import log_loss, accuracy_score, classification_report

print("Log Loss:", log_loss(y_test, y_pred_prob))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
