```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_excel('../Cases/Energy Efficiency/ENB2012_data.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.drop(['Y1', 'Y2'], axis=1)
y1 = df['Y1']
y2 = df['Y2']
```

### Ridge

```python
pipe = Pipeline([('preprocessing', StandardScaler()), ('model', Ridge())])
param_grid = {'model__alpha': np.linspace(0.1, 3, 10), 'preprocessing': [StandardScaler(), MinMaxScaler()]}

gcv = GridSearchCV(pipe, param_grid=param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=23), scoring='r2')

# Fit the model for Y1
gcv.fit(X, y1)
print(f"Best parameters for Ridge on Y1: {gcv.best_params_}")
print(f"Best score for Ridge on Y1: {gcv.best_score_}")

# Fit the model for Y2
gcv.fit(X, y2)
print(f"Best parameters for Ridge on Y2: {gcv.best_params_}")
print(f"Best score for Ridge on Y2: {gcv.best_score_}")
```

### Lasso

```python
pipe = Pipeline([('preprocessing', StandardScaler()), ('model', Lasso())])
param_grid = {'model__alpha': np.linspace(0.1, 3, 10), 'preprocessing': [StandardScaler(), MinMaxScaler()]}

gcv = GridSearchCV(pipe, param_grid=param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=23), scoring='r2')

# Fit the model for Y1
gcv.fit(X, y1)
print(f"Best parameters for Lasso on Y1: {gcv.best_params_}")
print(f"Best score for Lasso on Y1: {gcv.best_score_}")

# Fit the model for Y2
gcv.fit(X, y2)
print(f"Best parameters for Lasso on Y2: {gcv.best_params_}")
print(f"Best score for Lasso on Y2: {gcv.best_score_}")
```

### Elastic Net

```python
pipe = Pipeline([('preprocessing', StandardScaler()), ('model', ElasticNet())])
param_grid = {'model__alpha': np.linspace(0.001, 4, 20), 'model__l1_ratio': np.linspace(0.001, 0.999, 20), 'preprocessing': [StandardScaler(), MinMaxScaler()]}

gcv = GridSearchCV(pipe, param_grid=param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=23), scoring='r2')

# Fit the model for Y1
gcv.fit(X, y1)
print(f"Best parameters for ElasticNet on Y1: {gcv.best_params_}")
print(f"Best score for ElasticNet on Y1: {gcv.best_score_}")

# Fit the model for Y2
gcv.fit(X, y2)
print(f"Best parameters for ElasticNet on Y2: {gcv.best_params_}")
print(f"Best score for ElasticNet on Y2: {gcv.best_score_}")
```

### KNN

```python
pipe = Pipeline([('preprocessing', StandardScaler()), ('model', KNeighborsRegressor())])
param_grid = {'model__n_neighbors': np.arange(1,11), 'preprocessing': [StandardScaler(), MinMaxScaler()]}

gcv = GridSearchCV(pipe, param_grid=param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=23), scoring='r2')

# Fit the model for Y1
gcv.fit(X, y1)
print(f"Best parameters for KNeighborsRegressor on Y1: {gcv.best_params_}")
print(f"Best score for KNeighborsRegressor on Y1: {gcv.best_score_}")

# Fit the model for Y2
gcv.fit(X, y2)
print(f"Best parameters for KNeighborsRegressor on Y2: {gcv.best_params_}")
print(f"Best score for KNeighborsRegressor on Y2: {gcv.best_score_}")
```
